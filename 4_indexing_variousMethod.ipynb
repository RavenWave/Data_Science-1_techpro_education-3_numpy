## Selection on a Condition

```python
np.random.seed(40)
arr3 = np.random.randint(1, 50, size=(3,4))
arr3
```

```python
arr3 > 20
```

```python
arr3[arr3 > 20] # yukardaki True'lar bir array halinde yazdrdk
```

```python
arr3[1] > 10
```

```python
arr3[1][arr3[1] > 10]
```

```python
arr4 = np.arange(1, 51)
arr4
```

```python
arr4 % 2 == 0
```

```python
arr4[arr4 % 2 == 0]     # evens
```

```python
arr4[arr4 % 2 != 0]     # odds
```

```python
arr4[(arr4 % 2 == 0) == False]
```

```python
mask = np.array([1, 0, 1, 0, 0, 1], dtype=bool)
mask
```

```python
arr2[mask]
```

```python
arr2[mask, 2]
```

```python
arr2[2, mask]
```
